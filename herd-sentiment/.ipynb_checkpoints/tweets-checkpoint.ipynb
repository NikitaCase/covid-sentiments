```python
# Tweepy Setup
auth = tweepy.OAuthHandler(api_key, api_secret_key)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# Define some search terms
# Limited number of API calls so want to limit unusable data 

search_terms = ['covid+vaccine OR covid-19+vaccine OR pfizer OR pfizer-biontech OR moderna OR astrazeneca OR astra+zeneca -filter:retweets']
end_date = date(2021,3,4)
limit = 18000
end_date
```

```python
def GetTweets(end_date, search_terms, limit):
    data = []
    raw = []
    tweets = tweepy.Cursor(api.search, until=end_date, max_id ='1367261441515016193', location='canada', place='canada', lang ='en', q=search_terms, tweet_mode='extended').items(limit)
    for t in tweets: 
        tweet ={
            'created' :t.created_at,
            'id_str' : t.id_str,
            'text' : t.full_text,
            'user_id' :t.user.id,
            'screen_name': t.user.screen_name,
            'location': t.user.location,
            'followers_count': t.user.followers_count,
            'user_favourites_count': t.user.favourites_count,
            'time_zone' :t.user.time_zone,
            'geo_enabled' :t.user.geo_enabled,
            'verified' :t.user.verified,
            'status_count': t.user.statuses_count,
            'geo':t.geo,
            'coords':t.coordinates,
            'place':t.place,
            'retweet_count': t.retweet_count,
            'tweet_favourite_count' : t.favorite_count,
            'retweeted' : t.retweeted,
            'lang':t.lang
        }
    # add sorted data to one list 
        data.append(tweet)
    # keep raw data in case I realize later that I've removed something important 
        raw.append(t)
    tweets_df = pd.DataFrame(data)
    tweets_df.to_csv('tst.csv', index=False, encoding='UTF-8')
    with open('tst.json', 'w') as file:
        for resp in raw:
            jso = json.dumps(resp._json,sort_keys=True, indent=3)
            file.write(jso)
```

```python
def CleanFrame(df):
    df = df.drop_duplicates(subset='id_str')
    df['date'] = df['created'].str.split(' ').str[0]
    df = df.drop(columns=['created'])
    df['date'] = pd.to_datetime(df['date'])
    return df

    
    
```

```python
CleanFrame(df)
df.head()
```

```python
def CleanTweets(txt): 
    # remove @ username 
    # txt = re.sub(r'@[A-Za-z0-9)]+','', txt)
    
    # remove RT    
    txt = re.sub(r'RT[\s]+','',txt)
    
    # remove @ from username
    txt = txt.replace('@','')
    
    # remove # but leave txt
    txt = re.sub(r'#','', txt)
    
    # remove hyperlinks
    txt = re.sub(r'https?:\/\/\S+', '', txt)
    
    return txt
```

```python
df['text'] = df['text'].apply(CleanTweet)
df.head()
```

```python
# to get subjectivity 
def RateSubjectivity(txt):
    return TextBlob(txt).sentiment.subjectivity
```

```python
# to get polarity
def RatePolarity(txt):
    return TextBlob(txt).sentiment.polarity
```

```python
# Create Columns
df['Subjectivity'] = df['text'].apply(RateSubjectivity)
df['Polarity'] = df['text'].apply(RatePolarity)
df.head()
```
