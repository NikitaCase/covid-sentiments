```python
def Scrape_Tweets(location, filename):
    c = twint.Config()
    c.Since='2020-12-1'
    c.Until='2020-12-06'
    c.Near= location    
    c.Search='covid vaccine'
    c.Store_csv = True
    c.Output = filename
    
    twint.run.Search(c)    
```

```python
Scrape_Tweets('Vancouver', 'raw_covvac_vancouver.csv')
```

```python
Scrape_Tweets('Montreal','raw_covvac_montreal.csv')
```
